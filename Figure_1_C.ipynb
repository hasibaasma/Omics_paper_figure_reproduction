```r
# Load required libraries
library(DESeq2)
library(dplyr)
library(tibble)

# Read the HTSeq count data
#count_data <- read.csv("htseq-count.csv", row.names = 1)
count_data <- read.delim("GSE282695_htseq_count.tsv", sep="\t", header=TRUE, check.names=FALSE, stringsAsFactors=FALSE)

# Use the first column as row names, making them unique if necessary
rownames(count_data) <- make.unique(count_data$GENE)

# Remove the GENE column now used as row names
count_data <- count_data[, -1]

# Keep the first column as a regular column
#head(count_data)
all(count_data >= 0)
count_data[ count_data < 0, ]
# Print all column names to help with sample identification
cat("All column names in the count data:\n")
print(colnames(count_data))
cat("\n")

# Based on GEO metadata (GSE282695), there are 18 samples:
# CreER: S1, S2, S3 for each of: no LPS, LPS 1h, LPS 6h (9 samples total)
# CreER Stat3f/f: S1, S2, S3 for each of: no LPS, LPS 1h, LPS 6h (9 samples total)
#
# For Figure 1C, we need ONLY the PBS (no LPS) samples:
# - CreER S1, S2, S3 no LPS (3 samples)
# - CreER Stat3f/f S1, S2, S3 no LPS (3 samples)

# You need to manually identify which columns correspond to PBS-only samples
# Based on the pattern, it seems:
# S01, S02, S03 in column names likely correspond to biological replicates S1, S2, S3
# R01, R02, R03 might correspond to different treatment times

# PLEASE MANUALLY MAP YOUR COLUMN NAMES TO CONDITIONS:
# Update this section based on your actual column naming scheme
sample_names <- colnames(count_data)

# Create metadata dataframe
coldata <- data.frame(
  sample = sample_names,
  row.names = sample_names
)

# CRITICAL: Correct sample mapping based on diagnostic analysis
# The naming scheme is: F#_X##
# - F1, F2, F3 = biological replicates (3 replicates per condition)
# - R vs S = genotype (R = CreER control, S = CreER Stat3f/f)
# - 01, 02, 03 = treatment (01 = PBS, 02 = LPS 1h, 03 = LPS 6h)
#
# For Figure 1C (PBS only), we need:
# - CreER PBS: F1_R01, F2_R01, F3_R01
# - CreER Stat3f/f PBS: F1_S01, F2_S01, F3_S01

# Manually select the correct PBS samples
pbs_samples <- c("F1_R01", "F2_R01", "F3_R01",  # CreER PBS
                 "F1_S01", "F2_S01", "F3_S01")  # CreER Stat3f/f PBS

# Verify these samples exist in the data
pbs_samples <- pbs_samples[pbs_samples %in% sample_names]

cat("PBS-only samples selected for Figure 1C:\n")
print(pbs_samples)
cat("\n")

# Assign genotype based on R vs S prefix
coldata$genotype <- ifelse(grepl("_R", sample_names), "CreER", "CreER_Stat3f")
coldata$treatment <- ifelse(grepl("01$", sample_names), "PBS",
                           ifelse(grepl("02$", sample_names), "LPS_1h", "LPS_6h"))

cat("Sample assignments:\n")
print(table(coldata$genotype, coldata$treatment))
cat("\n")

# Subset the count data and metadata for PBS samples only
count_data_pbs <- count_data[, pbs_samples]
coldata_pbs <- data.frame(
  sample = pbs_samples,
  genotype = ifelse(grepl("_R", pbs_samples), "CreER", "CreER_Stat3f"),
  row.names = pbs_samples
)

# Convert to factor
coldata$genotype <- factor(coldata$genotype, levels = c("CreER", "CreER_Stat3f"))
coldata$treatment <- factor(coldata$treatment)

# Print sample information
cat("Sample information:\n")
print(table(coldata$genotype))
cat("\n")

# Create DESeq2 dataset using PBS samples only
dds <- DESeqDataSetFromMatrix(
  countData = count_data_pbs,
  colData = coldata_pbs,
  design = ~ genotype
)

# Filtering read counts (as mentioned in methods)
# Keep genes with at least 10 total reads across all samples
keep <- rowSums(counts(dds)) >= 10
dds <- dds[keep, ]

cat("Genes retained after filtering:", nrow(dds), "\n\n")

# Run DESeq2 analysis (normalization + differential expression)
# Uses negative binomial distribution followed by Wald statistical test
dds <- DESeq(dds)

# Get results comparing CreER Stat3f vs CreER (PBS treatment - baseline comparison)
# This identifies genes differentially expressed at homeostatic/baseline conditions
res <- results(dds, contrast = c("genotype", "CreER_Stat3f", "CreER"))

# Order by adjusted p-value (FDR)
res_ordered <- res[order(res$padj), ]

# Filter for significant DEGs
sig_genes <- res_ordered[which(res_ordered$padj < 0.05 & !is.na(res_ordered$padj)), ]

cat("Total significant DEGs (padj < 0.05):", nrow(sig_genes), "\n\n")

# Define the exact 40 genes from Figure 1C in the exact order shown
# Manually extracted from the high-resolution heatmap image
# These genes represent a curated list associated with pathogen defense
# and immune cell recruitment (many are interferon-stimulated genes)
# Note: The bottom ~10 genes show opposite pattern (higher in CreER, lower in Stat3f/f)
fig1c_genes <- c(
  "Fabp3", "Irgm2", "Oas3", "Ifi47", "Xaf1", "Cd300e", "Isg15", "Ifit3",
  "Irf7", "H2-Q6", "Olfr56", "Ifit2", "Oasl1", "Ifi44", "Klra17", "Ifi203-ps",
  "Slfn1", "Ifit3b", "Apold1", "Gprc5b", "Bst2", "Card14", "Abca5", "Cxcl9",
  "Bmpr1a", "Rab4a", "Ddit3", "Adam33", "Trim30c", "Wdfy1",
  "H1fx", "Caskin2", "Plxdc2", "Abca6", "Fcho1", "Ptgs1", "Klra3", "Stat3", "Abca9", "Ifi208"
)

cat("Attempting to extract", length(fig1c_genes), "genes from Figure 1C\n\n")

# Check which genes are present in the dataset
genes_present <- fig1c_genes[fig1c_genes %in% rownames(dds)]
genes_missing <- fig1c_genes[!fig1c_genes %in% rownames(dds)]

cat("Genes present in dataset:", length(genes_present), "\n")
if(length(genes_missing) > 0) {
  cat("Genes not found in dataset:\n")
  print(genes_missing)
  cat("\nAttempting to find alternative gene names...\n")

  # Try alternative gene names for genes with special characters
  for(gene in genes_missing) {
    # Try without "-ps" suffix
    if(grepl("-ps$", gene)) {
      alt_name <- sub("-ps$", "", gene)
      if(alt_name %in% rownames(dds)) {
        cat("Found", gene, "as", alt_name, "\n")
        genes_present <- c(genes_present, alt_name)
        # Replace in the original list
        fig1c_genes[fig1c_genes == gene] <- alt_name
      }
    }
    # Try alternative naming conventions
    if(grepl("-", gene)) {
      alt_name <- gsub("-", "", gene)
      if(alt_name %in% rownames(dds)) {
        cat("Found", gene, "as", alt_name, "\n")
        genes_present <- c(genes_present, alt_name)
        fig1c_genes[fig1c_genes == gene] <- alt_name
      }
    }
  }
  cat("\n")
}

# Use genes that are present (maintaining original order)
top40_genes <- fig1c_genes[fig1c_genes %in% rownames(dds)]

cat("Final gene count for analysis:", length(top40_genes), "\n\n")

# If you don't have the exact gene list, alternatively use top DEGs
# Uncomment these lines if the manual list doesn't work:
# cat("Using top 40 DEGs by adjusted p-value instead\n")
# top40_genes <- head(rownames(res_ordered), 40)

# Use regularized log transformation for visualization
# As stated in methods: "Regularized log-transformed data were used in visualization"
rld <- rlog(dds, blind = FALSE)
rld_counts <- assay(rld)

# Extract regularized log-transformed counts for the genes
top40_rld <- rld_counts[top40_genes, ]

# Separate samples by genotype for proper column ordering in heatmap
creer_pbs_samples <- pbs_samples[grepl("_R", pbs_samples)]
creer_stat3f_pbs_samples <- pbs_samples[grepl("_S", pbs_samples)]

# Sort samples to ensure consistent ordering (F1, F2, F3)
creer_pbs_samples <- sort(creer_pbs_samples)
creer_stat3f_pbs_samples <- sort(creer_stat3f_pbs_samples)

cat("CreER PBS samples:", length(creer_pbs_samples), "\n")
print(creer_pbs_samples)
cat("\nCreER Stat3f/f PBS samples:", length(creer_stat3f_pbs_samples), "\n")
print(creer_stat3f_pbs_samples)
cat("\n")

# Reorder columns: CreER PBS first, then CreER Stat3f PBS
ordered_samples <- c(creer_pbs_samples, creer_stat3f_pbs_samples)
top40_rld_ordered <- top40_rld[, ordered_samples]

# Scale the data by row (z-score normalization) for heatmap visualization
# This centers expression around 0 and scales to unit variance
top40_scaled <- t(scale(t(top40_rld_ordered)))

# Create output dataframe with gene names and scaled expression values
output_df <- as.data.frame(top40_scaled)
output_df <- rownames_to_column(output_df, var = "Gene")

# Add gene statistics (log2FoldChange, p-value, adjusted p-value)
gene_stats <- as.data.frame(res[top40_genes, c("baseMean", "log2FoldChange", "pvalue", "padj")])
gene_stats <- rownames_to_column(gene_stats, var = "Gene")

# Merge statistics with scaled expression values
output_df <- merge(gene_stats, output_df, by = "Gene")

# Reorder to match the original figure gene order
output_df <- output_df[match(top40_genes, output_df$Gene), ]

# Write to CSV
write.csv(output_df, "output.csv", row.names = FALSE)

# Print summary
cat("\nAnalysis complete!\n")
cat("Genes extracted and saved to output.csv:", nrow(output_df), "\n\n")
cat("Sample of genes and their statistics:\n")
print(output_df[1:min(10, nrow(output_df)), c("Gene", "log2FoldChange", "padj")])

# Optional: Create heatmap visualization to verify reproduction
# library(pheatmap)

# # Create annotation for sample groups
# annotation_col <- data.frame(
#   Genotype = c(rep("CreER PBS", length(creer_pbs_samples)),
#                rep("CreER Stat3P PBS", length(creer_stat3f_pbs_samples))),
#   row.names = ordered_samples
# )

# # Define color scheme similar to the paper
# color_scheme <- colorRampPalette(c("purple", "black", "yellow"))(100)

# # Generate heatmap matching Figure 1C
# pheatmap(
#   top40_scaled,
#   cluster_rows = FALSE,
#   cluster_cols = FALSE,
#   annotation_col = annotation_col,
#   color = color_scheme,
#   show_colnames = TRUE,
#   show_rownames = TRUE,
#   main = "Figure 1C - Top 40 DEGs (CreER Stat3f/f vs CreER PBS)",
#   fontsize_row = 7,
#   fontsize_col = 7,
#   border_color = NA
# )

library(pheatmap)

# Create annotation for sample groups
annotation_col <- data.frame(
  Genotype = c(rep("CreER PBS", length(creer_pbs_samples)),
               rep("CreER Stat3P PBS", length(creer_stat3f_pbs_samples))),
  row.names = ordered_samples
)

# Define color scheme matching the paper exactly
# The original Figure 1C uses the viridis color palette
library(viridis)

# Viridis palette matching the original (-2 to +2 scale)
color_scheme <- viridis(100, option = "D")

# Set breaks to match the scale shown in Figure 1C (-2, -1, 0, 1, 2)
color_breaks <- seq(-2, 2, length.out = 101)

# Generate heatmap matching Figure 1C
pheatmap(
  top40_scaled,
  cluster_rows = FALSE,
  cluster_cols = FALSE,
  annotation_col = annotation_col,
  color = color_scheme,
  breaks = color_breaks,
  show_colnames = TRUE,
  show_rownames = TRUE,
  main = "Figure 1C - Top 40 DEGs (CreER Stat3f/f vs CreER PBS)",
  fontsize_row = 7,
  fontsize_col = 7,
  border_color = NA,
  legend_breaks = c(-2, -1, 0, 1, 2),
  legend_labels = c("-2", "-1", "0", "1", "2")
)

cat("\nHeatmap visualization generated!\n")
cat("\nNote: The genes shown in Figure 1C are described in the paper as:\n")
cat("'a curated list of genes associated with pathogen defense and immune cell recruitment'\n")
cat("rather than simply the top 40 DEGs by statistical significance.\n")
```
